程式分兩部分，一個是擊中海底生物的函數，另一個是擊中空中海鷗的函數

* 擊中海底生物的功能 : 增加趣味性(擊中多樣性)，完全不影響遊戲公平

 * 鯨魚共 1 隻，面積為 3x4，隨機分佈

 * 水母共 30 隻，面積為 1x1，隨機分佈

* 擊中海鷗的功能 : 可對此次攻擊免疫，並通知攻擊者擊中未知目標物，可能是海鷗也有可能是船

 * 海鷗共 8 隻，面積為 1x1，隨機分佈

```python
%%writefile BattleShip.cpp

#include<iostream>
#include<cstdlib>
#include<ctime>
using namespace std;

void seaCreatures( int attackX, int attackY, int row, int column, int grid[][]){
    
  srand(time(0));

  int whaleX = 1+rand()%row, whaleY = 1+rand()%column;    //在海中佈置鯨魚(海底生物1)，共一隻鯨魚，面積為3x4

  whale_and_wall_spacing_x = whaleX + 3 - column - 1;
  whale_and_wall_spacing_y = whaleY + 4 - column - 1;

  if( whale_and_wall_spacing_x > 0 ){ //如果鯨魚的x邊長度超出牆

    whaleX -= whale_and_wall_spacing_x;

  }

  if( whale_and_wall_spacing_y > 0 ){ //如果鯨魚的y邊長度超出牆
                                    
    whaleY -= whale_and_wall_spacing_y;

  }

  for(int a = whaleX; a < whaleX + 3; a++){
      
    for(int b = whaleY; b < whaleY + 4; b++){
        
      if( grid[a][b] != 1){   //將鯨魚位置設置在非船位置上，若擊中則告訴攻擊方擊中鯨魚
        
        grid[a][b] = 2;    //將海底生物位置設為2

      }

      if( grid[attackX][attackY] == 2 ){  //當擊中位置為2時
          
        cout << "You hit the whale!! Say sorry to the whale!!" << endl;

      }

    }

  }


  for(int i=0; i<30; i++){   //在海中佈置水母(海底生物2)，共30隻，面積為1x1
      
    int jellyfishX = 1+rand()%row;
    int jellyfishY = 1+rand()%column;

    if( grid[jellyfishX][jellyfishY] != 1){
        
      grid[jellyfishX][jellyfishY] == 3;

    }

    if( grid[attackX][attackY] == 3 ){
        
      cout << "You hit a jellyfish" << endl;

    }

  }

}



void main( int attackX, int attackY, int row, int column){
    
  srand(time(0));

  int m, n, seagull[m][n];

  for(int k = 0; k < 8; k++){  //在空中隨機佈置海鷗位置，共8隻，體積為1x1
      
    m = 1+rand()%row;

    n = 1+rand()%column;

    seagull[m][n] = 1;

    if( seagull[attackX][attackY] == 1 ){   //確認攻擊情況(攻擊到海鷗會使該次攻擊無效化，並告訴攻擊方擊中未知目標，可能是船也可能是海鷗)
      
      cout << "You hit the seagull in the sky, and it makes your attack invalid!" << endl;

    }  //主程式必須在attack那部分增加一個前置if迴圈，判斷是否因攻擊到海鷗而抵擋此次攻擊

  }

}
```

Global variables

- int column : x 軸行數
- int row : y 軸行數
- int i : 第 i 行
- int j : 第 j 列
- int grid[column][row] : 矩陣
- int ship_length_x : 船在 x 向的長度
- int ship_length_y : 船在 y 向長度
- int ship_size : 船的總面積(佔幾個)
- int attackX : 對方攻擊的 x 座標
- int attackY : 對方攻擊的 y 座標
- int countOne : 我方基地中船還在的 grid 有幾格 One
